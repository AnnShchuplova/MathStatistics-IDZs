```python
# 3. Нахождение точечных оценок математического ожидания и дисперсии
def expectation(array):
    return sum(array) / len(array)

def variance(array):
    x_bar = expectation(array)
    return sum((x - x_bar) ** 2 for x in array) / (len(array) - 1)
mean = expectation(data)
var = variance(data)
print(f"Выборочное матожидание: {round(mean, 3)}")
print(f"Выборочная дисперсия (несмещенная): {round(var, 3)}")
```

4. Выдвижение гипотезы о виде закона распределения наблюдаемой случайной величины

Анализ результатов наблюдений показывает, что гистограмма имеет форму, похожую на нормальное распределение. Поэтому выдвигается гипотеза, что наблюдаемая случайная величина имеет нормальное распределение:

```python
# 5. Проверка гипотезы с помощью критерия хи-квадрат
def chi_square_test(data, alpha=0.05):
    n = len(data)
    k = int(1 + 3.322 * np.log10(n))
    min_val = min(data)
    max_val = max(data)
    interval_size = (max_val - min_val) / k
    intervals = [(min_val + i * interval_size, min_val + (i + 1) * interval_size) for i in range(k)]
    expected_frequencies = [len(data) * (stats.norm.cdf(intervals[i][1], mean, np.sqrt(var)) - stats.norm.cdf(intervals[i][0], mean, np.sqrt(var))) for i in range(k)]
    chi_square_stat = sum([(frequencies[i] - expected_frequencies[i])**2 / expected_frequencies[i] for i in range(k)])
    critical_value = stats.chi2.ppf(1 - alpha, k - 3)
    if chi_square_stat < critical_value:
        print("Гипотеза о нормальности распределения не отвергается.")
    else:
        print("Гипотеза о нормальности распределения отвергается.")

chi_square_test(data)
```

```python
# 6. Построение доверительных интервалов уровня надёжности 0.95 для математического ожидания и дисперсии
confidence_level = 0.95
alpha = (1 - confidence_level) / 2
n = len(data)
mean_confidence_interval = (np.mean(data) - stats.t.ppf(1 - alpha, n - 1) * np.sqrt(var) / np.sqrt(n), np.mean(data) + stats.t.ppf(1 - alpha, n - 1) * np.sqrt(var) / np.sqrt(n))
variance_confidence_interval = ((n - 1) * var / stats.chi2.ppf(1 - alpha, n - 1), (n - 1) * var / stats.chi2.ppf(alpha, n - 1))
print(f"Доверительный интервал для математического ожидания (уровень надёжности {confidence_level}):")
print(f"  Нижняя граница: {round(mean_confidence_interval[0], 3)}")
print(f"  Верхняя граница: {round(mean_confidence_interval[1], 3)}")
print()
print(f"Доверительный интервал для дисперсии (уровень надёжности {confidence_level}):")
print(f"  Нижняя граница: {round(variance_confidence_interval[0], 3)}")
print(f"  Верхняя граница: {round(variance_confidence_interval[1], 3)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Обратная функция распределения
def inverse_cdf(u):
    return 10 * np.sqrt(u)

# Генерация выборки объёма 500
np.random.seed(42)  # Для воспроизводимости
u = np.random.uniform(0, 1, 500)  # Равномерно распределенные числа на [0, 1]
samples_500 = inverse_cdf(u)  # Применяем обратную функцию

# Построение гистограммы
plt.hist(samples_500, bins=10, density=True)
plt.title("Гистограмма сгенерированной выборки")
plt.xlabel("Значения")
plt.ylabel("Плотность")
plt.show()
```

Метод обратного преобразования заключается в том, что функция распределения биективна и все значения из неё попадают в отрезок [0; 1], тогда можем найти обратную к этой функции и те значения, которые будут генерироваться равномерным распределением в этом же интервале, будут как раз задавать значение нужной выборки.

```python
display(Latex(f"H_0: \\quad Z_n \\sim F(x)")) # проверяем, что выборки соответствуют распределнию
display(Latex(f"H_A: \\quad Z_n \\nsim F(x)"))
```

```python
def chi_square_test(sample, k, alpha=0.05):
    # Разбиение на интервалы
    min_val = 0  # Минимальное значение распределения
    max_val = 10  # Максимальное значение распределения
    intervals = np.linspace(min_val, max_val, k + 1)  # Границы интервалов

    # Наблюдаемые частоты
    observed_freq, _ = np.histogram(sample, bins=intervals)

    # Ожидаемые частоты
    expected_freq = []
    for i in range(k):
        # Вероятность попадания в интервал [a, b)
        a = intervals[i]
        b = intervals[i + 1]
        prob = (b / 10)**2 - (a / 10)**2  # F(b) - F(a)
        expected_freq.append(prob * len(sample))

    # Вычисление статистики хи-квадрат
    chi_square_stat = np.sum((observed_freq - expected_freq)**2 / expected_freq)

    # Критическое значение
    degrees_of_freedom = k - 1  # Число степеней свободы
    critical_value = stats.chi2.ppf(1 - alpha, degrees_of_freedom)

    # Проверка гипотезы
    if chi_square_stat < critical_value:
        print(f"Гипотеза не отвергается (χ² = {round(chi_square_stat, 2)} < {round(critical_value, 2)}).")
    else:
        print(f"Гипотеза отвергается (χ² = {round(chi_square_stat, 2)} >= {round(critical_value, 2)}).")
```

```python
# Проверка для выборки объёма 50
print("Выборка объёма 50:")
chi_square_test(samples_50, k= 1 + int(3.322 * np.log10(50)))

# Проверка для выборки объёма 500
print("\nВыборка объёма 500:")
chi_square_test(samples_500, k= 1 + int(3.322 * np.log10(500)))
```

```python
def t_test(data, m0, alpha=0.05):
    n = len(data)
    mean = np.mean(data)  # Выборочное среднее
    std_dev = np.std(data, ddof=1)  # Выборочное стандартное отклонение

    # Вычисление t-статистики
    t_stat = (mean - m0) / (std_dev / np.sqrt(n))

    # Критическое значение
    degrees_of_freedom = n - 1  # Число степеней свободы
    t_critical = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)  # Двусторонний тест

    # p-значение
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), degrees_of_freedom))

    print(f"Выборочное среднее: {mean:.4f}")
    print(f"t-статистика: {t_stat:.4f}")
    print(f"Критическое значение: {t_critical:.4f}")
    print(f"p-значение: {p_value:.4f}")

    # Проверка гипотезы
    if abs(t_stat) < t_critical:
        print(f"Гипотеза H0 не отвергается (|t| = {round(abs(t_stat), 3)} < {round(t_critical, 3)}).")
    else:
        print(f"Гипотеза H0 отвергается (|t| = {round(abs(t_stat), 3)} >= {round(t_critical, 3)}).")
```

```python
# Проверка для выборки объёма 50
print("Выборка объёма 50:")
t_test(samples_50, 6.5)

# Проверка для выборки объёма 500
print("\nВыборка объёма 500:")
t_test(samples_500, 6.5)
```

На основании полученных результатов можно сделать вывод, что как выборка объема 50, так и выборка объема 500 соответствуют нашему распределению. Также, гипотезы о том, что математическое ожидание наблюдаемой СВ равно m0 = 6.5 не отвергаются для обеих выборок.

Можно предположить, что при большей выборке гипотеза могла бы быть отвергнута, так как можно точнее предсказать математическое ожидание, и увидеть, что наше предположение всё менее вероятно.
